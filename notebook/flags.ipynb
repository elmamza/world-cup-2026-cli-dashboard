```python
from PIL import Image
import requests
from io import BytesIO

GRID_W = 25
GRID_H = 14

def rgb_to_hex(rgb):
    return '#%02x%02x%02x' % rgb

url = "https://flagcdn.com/w320/se.png"

response = requests.get(url)
img = Image.open(BytesIO(response.content)).convert("RGB")

# resize direct en grille
img = img.resize((GRID_W, GRID_H), Image.NEAREST)

pixels = list(img.getdata())

print("{")
for y in range(GRID_H):
    row = []
    for x in range(GRID_W):
        pixel = pixels[y * GRID_W + x]
        row.append(f'"{rgb_to_hex(pixel)}"')
    print("    {" + ", ".join(row) + "},")
print("}")
```

```python
from PIL import Image
import requests
from io import BytesIO

GRID_W = 25
GRID_H = 14

fifa_to_iso2 = {
    "RSA": "za",
    "SCO": "gb",
    "HAI": "ht",
    "PAR": "py",
    "TUR": "tr",
    "CZE": "cz",
    "BIH": "ba",
    "CUW": "cw",
    "CIV": "ci",
    "EGY": "eg",
    "NZL": "nz",
    "CPV": "cv",
    "IRQ": "iq",
    "NOR": "no",
    "ALG": "dz",
    "AUT": "at",
    "JOR": "jo",
    "COD": "cd",
    "UZB": "uz",
    "COL": "co",
    "PAN": "pa",
}

def rgb_to_hex(rgb):
    return '#%02x%02x%02x' % rgb

def fetch_flag(iso2):
    url = f"https://flagcdn.com/w320/{iso2}.png"
    r = requests.get(url)
    img = Image.open(BytesIO(r.content)).convert("RGB")
    img = img.resize((GRID_W, GRID_H), Image.NEAREST)
    return img

def img_to_grid(img):
    pixels = list(img.getdata())
    grid = []
    for y in range(GRID_H):
        row = []
        for x in range(GRID_W):
            pixel = pixels[y * GRID_W + x]
            row.append(rgb_to_hex(pixel))
        grid.append(row)
    return grid

print("var additionalFlags = map[string][14][25]string{")

for fifa, iso in fifa_to_iso2.items():
    try:
        img = fetch_flag(iso)
        grid = img_to_grid(img)

        print(f'    "{fifa}": {{')
        for row in grid:
            print("        {" + ", ".join(f'"{c}"' for c in row) + "},")
        print("    },")
    except Exception as e:
        print(f"// ERROR {fifa} ({iso}): {e}")

print("}")
```
